```python
import pandas as pd
import numpy as np
```

```python
dfabundance = pd.read_csv('Data/abundance', header=0, index_col=0, sep='\t')
dfcount = pd.read_csv('Data/count', header=0, index_col=0, sep='\t')
```

```python
def extract_Campylobacterales(df):
    
    dfcamp=df[df['taxonomy'].str.contains('Campylobacterales')]

    genus = dfcamp["taxonomy"].str.split(";", expand=True).iloc[:,0:7]
    genus.columns = ["domain","phylum","class","order","family","genus","species"]

    dfcamp_tax = pd.concat([dfcamp, genus["genus"]], axis=1, sort=False)
    dfgroupby=dfcamp_tax.groupby('genus').sum()

    dfgroupby.loc['Proteobacteria; Epsilonproteobacteria; Campylobacterales']=dfgroupby.sum()
    
    return dfgroupby
```

```python
dfg_abundance=extract_Campylobacterales(dfabundance)
dfg_count=extract_Campylobacterales(dfcount)
```
